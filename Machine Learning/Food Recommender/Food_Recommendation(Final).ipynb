```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df.info()
```

```python
user_ratings = df.pivot_table(index=["UserId"],columns=["Item Name"],values="Ratings")
user_ratings = user_ratings.dropna(thresh=10,axis=1).fillna(0,axis=1) 
user_ratings.head()
```

```python
item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df.head()
```

### Making Recommendation

```python
def get_similar_food(food_name,user_rating):
    similar_score = item_similarity_df[food_name]*(user_rating-2.5) ## if the rating is below 3 then subtracting by mean rating i.e 2.5 will push them to negative side and vice versa.
    similar_score = similar_score.sort_values(ascending=False)
    
    return similar_score
```

```python
food_name_ratings = [("Prawn Karahi",1)] ## Food Name and our rating
similar_foods = pd.DataFrame()

for food,rating in food_name_ratings:
    similar_foods = similar_foods.append(get_similar_food(food,rating),ignore_index=True)
```

```python
similar_foods.sum().sort_values(ascending=False).head(10)
```
